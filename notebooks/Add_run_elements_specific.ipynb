```python
import yaml
import pandas as pd
import numpy as np
import itertools

def generate_model_template(
    dataset_name, model_name, best_params=None, param_grid=None
):
    template = {
        "dataset": dataset_name,
        "model": model_name,
        "best_params": {},
        "param_grid": {},
    }

    if best_params:
        template["best_params"] = best_params

    if param_grid:
        template["param_grid"] = param_grid

    return template
```

```python
import math

def balanced_partitions(values_dict, n_parts):
    # Get total sum of values
    total_sum = sum(values_dict.values())
    
    # Calculate ideal sum per part
    ideal_sum = total_sum / n_parts
    
    # Initialize partitions lists
    partitions = [[] for _ in range(n_parts)]
    
    # Sort keys by value (descending)
    sorted_keys = sorted(values_dict, key=lambda k: values_dict[k], reverse=True)  
    
    # Add keys to partitions greedily
    current_sum = [0] * n_parts
    for key in sorted_keys:
        min_idx = current_sum.index(min(current_sum))
        partitions[min_idx].append(key)
        current_sum[min_idx] += values_dict[key]
    
    # Return partitions list    
    return partitions

n = 3  # Number of parts
result = balanced_partitions(initial_queue_weights, n)
print(result)
```
