```python
import yaml

def generate_model_template(dataset_name, model_name, best_params = None, param_grid = None):
    
    template = {
                "dataset": dataset_name,
                "model": model_name,
                "best_params": {},
                "param_grid": {}
    }
    

    if best_params:
        template["best_params"] = best_params
        
    if param_grid:
        template["param_grid"] = param_grid

    return template
```
